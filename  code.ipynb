```python
4. Data Visualization

A.Distribution of Calories

the histogram and boxplot for the said distribution can be plotted together with help of a subplt,so that we can compare both 
the visuals side by side to get more comprehensive insights about the distribution as well as the outliers.
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(mcd_data['Calories'],  bins=10,  kde=True)
plt.title('Histogram of Calorie Distribution')
plt.xlabel('Calories')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.boxplot(x=mcd_data['Calories'])
plt.title('Box plot of Calorie Distribution')
plt.xlabel('Calories')

plt.tight_layout()
plt.show()
```

```python
The shape of the histograam is right-skewed,with a peak at approximately 125-500 calories and a long toward the higher values
of calories.the majority of menu items fall on the lower end of the calorie count,and there are only a few high-calorie outliers
The observation are confirmed by the box plt, where the median is approximately 340 calories and the 25th and 75th percentiles 
are 210 and 500,respectively with high calories, starting from the " McFlurry with Reese's peanut Cups(Medium)" at 810 calories.
```
